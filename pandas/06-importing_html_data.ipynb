### Importing Data From a Webpage
The `pandas.read_html()` function parses HTML strings, files, or URLs to look for `<table>` tags and automatically extracts them into a list of pandas `DataFrame` objects. 

`read_html()` requires underlying parsing engines. Install them via your terminal. `pip install lxml html5lib beautifulsoup4`


#### Image:
![Image of a Webpage](../data/images/html-page.png)

```python
# Basic Import

# This is a w3schools page with data tables
url = "https://www.w3schools.com/html/html_tables.asp"

# Read all tables into a list
tables = pd.read_html(url)

# Grab the very table from that list using their indexs
df = tables[0]
df.head()
```

### 1. `storage_options` Parameter
The `storage_options` parameter accepts a dictionary of connection configurations to pass connection-specific arguments—like custom HTTP headers, authentication tokens, or cloud storage credentials—when reading an HTML file from a remote URL.

```python
# 1.1 Adding HTTP Headers (Bypassing Scrapers Blocks)

## Websites often block default Python scripts to prevent web scraping and give `HTTPError: HTTP Error 403: Forbidden` error.
## You need to trick those websites into thinking your Python script is a normal web browser (like Chrome or Firefox).

url = "https://www.worldometers.info/world-population/population-by-country/"

# Specifying a standard web browser user-agent string
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"
}

# Passing the headers via storage_options
tables = pd.read_html(url, storage_options=headers)

df = tables[0]
df.head()
```

### 2. `match` Parameter
The `match` parameter filters the extracted HTML tables and only returns those containing specific text or matching a regular expression. 

By default, `read_html()` parses every single table on a webpage, which can return dozens of unwanted DataFrames. The match argument acts as a precise filter to fetch exactly what you need.

#### Image:
![Image of a Webpage](../data/images/read-html-match1.png)

```python
# 2.1 Matching a Simple String
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"
}

url = "https://en.wikipedia.org/wiki/Minnesota"

## This only grabs tables containing the exact header text
tables = pd.read_html(url, storage_options=headers, match="Average daily maximum and minimum temperatures for selected cities in Minnesota")

## Even with 'match', a list is returned. Use [0] to grab the first matching DataFrame.
target_df = tables[0]

print(f"Found {len(tables)} table(s) matching the pattern.")
target_df.head()
```

#### Images:
![Image of a Webpage](../data/images/read-html-match2.png)
![Image of a Webpage](../data/images/read-html-match3.png)

```python
# 2.2 Using Regular Expressions (RegEx)
import re
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"
}

url = "https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)"

# Match tables that mention "GDP" and "country" while ignoring case sensitivity
regex_pattern = re.compile(r"gdp.*(country|region)", re.IGNORECASE)
tables = pd.read_html(url, storage_options=headers, match=regex_pattern)

target_df1 = tables[0]
target_df2 = tables[1]

print(f"Found {len(tables)} table(s) matching the pattern.")
target_df1.head()
```
