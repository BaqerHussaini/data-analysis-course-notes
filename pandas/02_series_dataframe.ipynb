## **Series**
---

### What Series is
A Pandas Series is a one-dimensional labeled array capable of holding any data type, such as integers, floats, strings, or Python objects. Conceptually, it is like a single column in an Excel spreadsheet or a database table.

### Creating Series
To create a Pandas Series, use the `pd.Series()` constructor. Series can be build from various data sources like Python lists, NumPy arrays, dictionaries, or scalar values.

```python
# 1. Creating an empty Series
my_series = pd.Series()
my_series
```

```python
# 2. From a Python List or Tuple
data_list = ["Apple", "Banana", "Oringe"]
my_series = pd.Series(data_list)
my_series
```

```python
# 3. From a Dictionary (Keys become the index labels)
my_series = pd.Series(data={"a": 200, "b": 400, "c": 500})
my_series
```

```python
# 4. From a Scalar (Repeats value across the given index)
my_series = pd.Series(10, index=[x for x in "abcd"])
my_series
```

```python
# 5. From a Python fuction (range)
my_series = pd.Series(range(50, 75, 5))
my_series
```

```python
# 6. From a Numpy Array
import numpy as np
np_data = np.array([1.4, 3.4, 8])
my_series = pd.Series(np_data)
my_series
```

```python
# 7. From a DataFrame
data = {
    "Name": ["Ali", "Mohammad", "Sadiq"],
    "Age": [22, 30, 15]
} 
user_data = pd.DataFrame(data)
user_names = pd.Series(user_data["Name"])
user_names
```

## __Common Series Creation Customizations__
---

### 1. `index` Parameter
The `index` parameter specifies the axis labels used to identify each element within the one-dimensional array. 

By default, if no index is passed, pandas automatically assigns a sequential integer index starting from `0`.

```python
# 1.1 Default Index
my_series = pd.Series([4, 9, 8, 4, 2])
my_series
```

```python
# 1.2 Custom Indexs Using a Python List, Tuple or Numpy Array
index_list = ["A", "B", "C", "D", "F"]
my_series = pd.Series([4, 9, 8, 4, 2], index=index_list)
my_series
```

```python
# 1.3 Custom Indexs Using range() Function
my_series = pd.Series([4, 9, 8, 4, 2], index=range(10, 20, 2))
my_series
```

```python
# 1.4 Giving Index with Data
dataset = {
    "Name": "Ahmad",
    "Age": 25,
    "Country": "Afghanistan"
}
my_series = pd.Series(dataset)
my_series
```

### 2. `dtype` Parameter
The `dtype` parameter explicitly sets the data type of the underlying array. If you do not specify this parameter, Pandas will automatically infer the data type based on the values provided in your dataset.

```python
# 2.1 Setting a Data Type
my_series = pd.Series(["1", "8", "9"], dtype=int)
my_series.dtypes
```

```python
# 2.2 Optimizing Memory Usage
my_series = pd.Series(["22", "83", "11"], dtype="int16")
my_series.dtype
```

### 3. `name` Parameter
The `name` parameter allows you to assign a human-readable label or identifier to the `Series` object. It serves as a core piece of metadata that dictates how the `Series` interacts with other parts of the pandas library.

Why use the name parameter?
* __DataFrame Conversion:__ When converting a `Series` into a `DataFrame`, the name automatically becomes the column header.
* __Automatic Extraction:__ If you pull a single column out of a `DataFrame`, pandas automatically populates the name attribute of that resulting `Series` with the original column name.
* __Plotting & Display:__ The name is automatically utilized as a default axis or legend label when using the `Series.plot()` method. It also prints cleanly at the bottom of the `Series` terminal output.

```python
# 3.1 Defining a name
dataset = {
    "Name": "Ahmad",
    "Age": 25,
    "Country": "Afghanistan"
}
info = pd.Series(dataset, name="User Info")
info
```

```python
# 3.2 Converting to a DataFrame
dataset = {
    "Name": "Ahmad",
    "Age": 25,
    "Country": "Afghanistan"
}
info = pd.Series(dataset, name="User Info")
user_data = pd.DataFrame(info)
user_data
```

```python
# 3.3 Extracting Series from a DataFrame
data = {
    "Name": ["Ali", "Mohammad", "Sadiq"],
    "Age": [22, 30, 15]
} 
user_data = pd.DataFrame(data)
user_names = pd.Series(user_data["Name"])
user_names
```

## **DataFrame**
---

### What DataFrame is
A Pandas DataFrame is a two-dimensional, tabular data structure in Python that organizes data into rows and columns, much like an Excel spreadsheet or a SQL database table. It can store different types of data such as numbers, text and dates across its columns.

### Creating DataFrame
To create a pandas DataFrame, use the `pd.DataFrame()` constructor. Pandas allows us to create a DataFrame from many data sources. We can create DataFrames directly from Python objects like lists and dictionaries or by reading data from external files like CSV, Excel or SQL databases.

```python
# 1. Creating an empty DataFrame
my_dataframe = pd.DataFrame()
print(my_dataframe)
```

```python
# 2. From Python Lists or Tuple
names = ["Ali", "Mohammad", "Sadiq", "Hassan"]
ages = [22, 25, 18, 10]
users_info = pd.DataFrame(data={"Name": names,"Age": ages})
users_info
```

```python
# 3. From a Python Dictionary
dataset = {
    "Name":["Ali", "Mohammad", "Hassan"],
    "Age": [22, 18, 25],
    "City": ["Kabul", "Bamyan", "Herat"]
}
users_info = pd.DataFrame(data=dataset)
users_info
```

```python
# 4. From a Numpy Array
data = {
    "A": np.array([1, 2, 3]),
    "B": np.array([4, 5, 6]),
    "C": np.array([7, 8, 9])
}
np_dataframe = pd.DataFrame(data)
np_dataframe
```

## __Common DataFrame Creation Customizations__
---

### 1. `index` Parameter
The `index` parameter specifies the row labels for the resulting DataFrame object. If you do not explicitly pass this parameter, pandas automatically generates a default numeric sequence starting from `0` using `pandas.RangeIndex`.

```python
# 2.1 Default Index
my_dataframe = pd.DataFrame({"A": [98, 34, 67, 8], "B": [22, 54, 82, 19], "C": [12, 90, 32, 43]})
my_dataframe
```

```python
# 2.2 Custom Indexs Using List
data = {
    "A": [98, 34, 67, 8],
    "B": [22, 54, 82, 19],
    "C": [12, 90, 32, 43]
}
indexs = ["CA", "NY", "AZ", "TX"]
my_dataframe = pd.DataFrame(data, index=indexs)
my_dataframe
```

```python
# 2.3 Custom Index Using range() Function
data = {
    "A": [98, 34, 67, 8],
    "B": [22, 54, 82, 19],
    "C": [12, 90, 32, 43]
}
my_dataframe = pd.DataFrame(data, index=range(10, 14))
my_dataframe
```

### 2. `columns` Parameter
The `columns` parameter explicitly sets or filters the column labels/names of your data table when initializing a new DataFrame.

```python
# 1. Data with the Column Labels
data = {
    "Name": ["Ali", "Hassan", "Mohammad"],
    "Age": [24, 18, 22],
    "City": ["Kabul", "Bamyan", "Herat"]
}
std_info = pd.DataFrame(data)
std_info
```

```python
# 2. Specifying Column Order
data = {
    "Name": ["Ali", "Hassan", "Mohammad"],
    "Age": [24, 18, 22],
    "City": ["Kabul", "Bamyan", "Herat"]
}
std_info = pd.DataFrame(data, columns=["Name", "City", "Age"])
std_info
```

```python
# 3. Adding Empty Column(s) to Existing Columns
data = {
    "Name": ["Ali", "Hassan", "Mohammad"],
    "Age": [24, 18, 22],
    "City": ["Kabul", "Bamyan", "Herat"]
}
std_info = pd.DataFrame(data, columns=["Name", "Last Name", "Age", "City"])
std_info
```

```python
# 4. Defining Columns During DataiFrame Creation
data =[
    ["Ali", 24, "Kabul"],
    ["Hassan", 18, "Bamyan"],
    ["Mohammad", 22, "Herat"]
]
std_info = pd.DataFrame(data, columns=["Name", "Age", "City"])
std_info
```

```python
# 5. Filtering Existing Data Structures
data = {
    "Name": ["Ali", "Hassan", "Mohammad"],
    "Age": [24, 18, 22],
    "City": ["Kabul", "Bamyan", "Herat"]
}
std_info = pd.DataFrame(data, columns=["Name", "Age"])
std_info
```

```python
# 6. Creates an Empty DataFrame with Specific Columns
std_info = pd.DataFrame(columns=["STD_ID", "STD_Score", "STD_State"])
std_info
```
