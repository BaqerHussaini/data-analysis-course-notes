### 1. `lines` Parameter
The `lines` parameter controls whether the input file should be read as a line-delimited JSON format. By default, `lines=False`. When you set `lines=True`, pandas treats each individual line in the file as a distinct, valid JSON object rather than parsing the entire file as a single massive JSON array or dictionary.

#### Image:
![Image of a Json File](../data/images/simple-json-file.png)

````python
# 1. Standard JSON Array (lines=False)
car_sales = pd.read_json("../data/json-files/car-sales.json")
car_sales.head(3)
````

#### Image:

````python
# 2. JSON Lines / NDJSON (lines=True)
house_price = pd.read_json("../data/json-files/house-price-parquet.json", lines=True)
house_price.head(3)
````

### 2. `orient` Parameter
The `orient` parameter specifies the expected layout or structure of the incoming JSON string or file so that pandas can correctly map it into a `DataFrame` or `Series`. By default, if `typ='frame'`, pandas assumes `orient='columns'`.

#### 1. `orient='records'` (Most Common)
Expects a list of dictionaries, where each dictionary represents a row with column-value pairs.
* __Best for:__ Standard web APIs and general data transfer.
* __JSON Structure:__
  ```
                  [
                    {"Name": "Alice", "Age": 25},
                    {"Name": "Bob", "Age": 30}
                  ]
  ```

#### 2. `orient='columns'` (Default)
Expects a nested dictionary structured as {#column -> {#index -> value}}.
* __Best for:__ Preserving row indexes without repeating column names for every row.
* __JSON Structure:__
  ```
                  {
                    "Name": {"0": "Alice", "1": "Bob"},
                    "Age": {"0": 25, "1": 30}
                  }
  ```

#### 3. `orient='index'`
Expects a nested dictionary structured as {#index -> {#column -> value}}.
* __Best for:__ Key-value lookups where the primary keys are the row labels.
* __JSON Structure:__
  ```
                  {
                    "0": {"Name": "Alice", "Age": 25},
                    "1": {"Name": "Bob", "Age": 30}
                  }
  ```

#### 4. `orient='split'`
Expects a dictionary with explicit, separate keys for columns, index, and data arrays.
* __Best for:__ Maximizing space efficiency while maintaining explicit axis names.
* __JSON Structure:__
  ```
                  {
                    "columns": ["Name", "Age"],
                    "index": [0, 1],
                    "data": [
                         ["Alice", 25],
                         ["Bob", 30]
                      ]
                  }
  ```

#### 5. `orient='values'`
Expects a bare array of arrays containing only the raw data values. No index or column labels are preserved.
* __Best for:__ Highly compressed data transfers where schemas are documented elsewhere.
* __JSON Structure:__
  ```
                  [
                    ["Alice", 25],
                    ["Bob", 30]
                  ]

  ```

#### 6. `orient='table'`
Expects a dictionary containing a schema field describing the data types and a data field formatted like records.
* __Best for:__ Safely transferring data between pandas sessions while keeping exact data types intact.
* __JSON Structure:__
  ```
                  {
                    "schema": {
                      "fields": [
                        {"name": "index", "type": "integer"},
                        {"name": "Name", "type": "string"},
                        {"name": "Age", "type": "integer"}
                      ],
                      "primaryKey": ["index"],
                      "pandas_version": "1.4.0"
                    },
                    "data": [
                      {"index": 0, "Name": "Alice", "Age": 25},
                      {"index": 1, "Name": "Bob", "Age": 30}
                    ]
                  }
  ```

### 4. `convert_dates` Parameter
The `convert_dates` parameter controls whether date-like strings or Unix timestamps in your JSON data are automatically parsed into Pandas `datetime64[ns]` objects.

````python
# 1. Default Behavior (True)
erp_sales_report = pd.read_json("../data/json-files/ERP_sales_report_east_20260607_084932.json")
erp_sales_report.head(3)
````

````python
erp_sales_report.dtypes
````

````python
# 2. Disabling Date Conversion (False)
erp_sales_report = pd.read_json("../data/json-files/ERP_sales_report_east_20260607_084932.json",
                                convert_dates=False)
erp_sales_report.head(3)
````

````python
erp_sales_report.dtypes
````

````python
# If we set keep_default_dates=False, read_json fuction stops to search for date columns by its own.
erp_sales = pd.read_json("../data/json-files/ERP_sales_report_east_20260607_084932.json", 
                         keep_default_dates=False)
erp_sales.head(3)
````

````python
erp_sales.dtypes
````

````python
# 3. Specifying Target Column(s)
erp_sales = pd.read_json("../data/json-files/ERP_sales_report_east_20260607_084932.json", 
                         keep_default_dates=False, convert_dates=["Date"])
erp_sales.head(3)
````

````python
erp_sales.dtypes
````
