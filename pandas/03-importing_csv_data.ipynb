### 1. `sep` Parameter
the `sep` parameter stands for "separator" (or delimiter) and is used to specify the character or regular expression that splits the columns in your text file. 

By default, `sep=','` because standard Comma-Separated Values (CSV) files use a comma to separate data fields, but you can change it to `sep=';'`, `sep='\t'` or others depending on you datasets.

#### Image:
![Image of a Commas Separated CSV File](../data/images/csv-file.png)

```python
# Default (,)
car_sales = pd.read_csv("../data/csv-files/car-sales.csv")
car_sales.head(3)
```

### 2. `header` Parameter
The `header` parameter determines which row (or rows) of your CSV file will be used as the DataFrame's column names. 

The default is `header='infer'`, but you can pass `None`, a integer number or a list of integers.

#### Image:
![Image of a CSV File with Header](../data/images/csv-file.png)

```python
# 2.1 Default Behavior (header='infer' <=> header=0)
car_sales = pd.read_csv("../data/csv-files/car-sales.csv")
car_sales.head(3)
```

```python
# 2.1.1 Default Behavior (header='infer' <=> header=0)
car_sales = pd.read_csv("../data/csv-files/car-sales.csv", skiprows=1)
car_sales.head(3)
```

```python
# 2.1.2 Default Behavior (header='infer' <=> header=None because names are given)
created_names = ["A", "B", "C", "D", "E"]
car_sales = pd.read_csv("../data/csv-files/car-sales.csv", names=created_names)
car_sales.head(3)
```

```python
# If your file has headers but you want to replace them with your own names, pass header=0 alongside your custom names list.
# This overrides the original text rather than pushing it down into the first data row.

car_sales = pd.read_csv("../data/csv-files/car-sales.csv", header=0, names=created_names)
car_sales.head(3)
```

```python
# 2.2 CSV File Has No Headers (header=None)
car_sales = pd.read_csv("../data/csv-files/car-sales.csv", skiprows=1, header=None,)
car_sales.head(3)
```

#### Image:
![Image of Multi Level Columns CSV File](../data/images/multi-level-columns-csv.png)

```python
# 2.3 Handling MultiIndex Headers (header=[int, int, ...])
# Use when your CSV file contains multi-level (hierarchical) column headers.
multi_level_columns = pd.read_csv("../data/csv-files/multi-level-columns.csv", header=[0, 1])
multi_level_columns
```

### 4. `index_col` Parameter
`index_col` parameter specifies which column(s) to use as the row labels (index) of the resulting DataFrame. 

By default `index_col=None`,  which prompts `pandas` to generate an automatic integer index (0, 1, 2, ...). Other things that you can pass are, `False`, `0`, a string(name) or a `list`.

### 5. `skiprows` Parameter
Allows you to exclude specific lines or blocks of data from being parsed when loading a CSV file. 

By default `skiprows=None`, but you can pass an integer, a list of integers or a callable/lambda fuction.

#### Image:
![Image of a CSV File](../data/images/read-csv-skiprows1.png)

```python
# 5.1 An Integer (Skip N Rows from Start)
heart_disease = pd.read_csv("../data/csv-files/heart-disease.csv", sep=";", skiprows=2)
heart_disease.head(3)
```

#### Image:
![Skip Specific Positions Image](../data/images/read-csv-skiprows2.png)

```python
# 5.2 A List of Integers (Skip Specific Row Indices)
car_sales_missing = pd.read_csv("../data/csv-files/car-sales-missing-data.csv", sep="\t", skiprows=[0, 3, 7])
car_sales_missing
```

```python
# 5.3 A Callable / Lambda Function (Conditional Skipping)
# 5.3.1 Checking Against an Outside List
skipped_records = [0, 3, 7]
car_sales_missing = pd.read_csv("../data/csv-files/car-sales-missing-data.csv", sep="\t", 
                                skiprows=lambda x: x in skipped_records)
car_sales_missing
```

```python
# 5.3.2 Checking a Pattern (Using Math)
car_sales_extended = pd.read_csv("../data/csv-files/car-sales-extended.csv", skiprows=lambda x: x%5==0)
car_sales_extended.head(3)
```

```python
# 5.3.3 Random Sampling (Using Probability)
import random
car_sales_extended = pd.read_csv("../data/csv-files/car-sales-extended.csv", 
                                 skiprows=lambda x: x>0 and random.random()>0.8)
car_sales_extended.head(3)
```

### 7. `usecols` Parameter
`usecols` parameter allows you to select and load only specific columns from a CSV file into memory. This is highly useful for reducing memory usage and speeding up file parsing when dealing with massive datasets. 

You can pass a list of column names, a list of integer indices, or a custom lambda function to the parameter.

#### Image:
![Image of a CSV File](../data/images/csv-file.png)

```python
##### 7.1 Column Names (String List)
car_sales = pd.read_csv("../data/csv-files/car-sales.csv", usecols=["Make", "Price"])
car_sales.head(3)
```

```python
# 7.2 Position Index (Integer List)
car_sales = pd.read_csv("../data/csv-files/car-sales.csv", usecols=[0, 4])
car_sales.head(3)
```

```python
# 7.3 Dynamic Filtering (Callable/Lambda)
# 7.3.1 Black-listing
ignored_cols = ["Make", "Odometer (KM)"]
car_sales = pd.read_csv("../data/csv-files/car-sales.csv", usecols=lambda col: col not in ignored_cols)
car_sales.head(3)
```

#### Image:
![Image of a CSV File](../data/images/read-csv-usecols.png)

```python
# 7.3.2 Pattern Matching (Partial Names)
erp_sales_report = pd.read_csv("../data/csv-files/ERP_sales_report_east_20260607_084932.csv", skiprows=3,
                               sep=";", usecols=lambda col: any(x in col.lower() for x in ["id", "name", "price"]))
erp_sales_report.head(3)
```

### 9. `dtype` Parameter
The `dtype` parameter allows you to explicitly force specific data types for your columns while importing a CSV file. 

By default, pandas tries to guess (infer) data types automatically, but this can lead to high memory consumption or incorrect data type assignments (like leading zeros being dropped from postal codes or ID numbers).

#### Image:
![Image of a CSV File](../data/images/csv-file.png)

```python
# The original data types of this CSV file
car_sales = pd.read_csv("../data/csv-files/car-sales.csv")
car_sales.info()
```

### 11. `true_values` Parameter
It is used to specify a list of additional strings that should be treated as the boolean `True`. 

By default, pandas only recognizes case-insensitive variations of 'True' (such as "true", "True", "TRUE") as boolean `True`. If your dataset uses alternative terms like "Yes", "Y", or "Active", you use this parameter alongside `false_values` to parse them cleanly into actual booleans during data import.


### 12. `false_values` Parameter
It is used to specify a list of additional strings that should be parsed as the boolean `False`.

By default, pandas only recognizes case-insensitive variations of 'False' (such as "false", "FALSE", "False") as boolean `False`. If your dataset uses alternative representations like "No", "N", "No", or "0", you use this parameter alongside `true_values` to parse them cleanly into actual booleans during data import.

#### Image:
![Image of a CSV File](../data/images/read-csv-true-values-false-values.png)

```python
# Define your custom True representations
students_payment_report = pd.read_csv("../data/csv-files/students-payment-report.csv",
                                     true_values=["Active", "Yes", "TRUE"],
                                     false_values=["Inactive", "No", "FALSE"]
                                     )
students_payment_report.dtypes
```

### 13. `keep_default_na` Parameter
`keep_default_na` parameter in controls whether built-in null identifiers (like "`NA`", "`NaN`", "`null`", or blank cells) are automatically converted into `NaN` values. It defaults to `True`.

#### Default Missing Values in Pandas
If you do not change `keep_default_na`, the following built-in strings are automatically parsed as `NaN`:
* `""` (empty strings), `NaN`, `nan`, `null`, `NULL`
* `N/A`, `n/a`, `NA`, `#N/A`, `#NA`
* `-NaN`, `-nan`, `1.#IND`, `-1.#IND`, `1.#QNAN`, `-1.#QNAN`

#### Image:
![Quick Breakdown of keep_default_na Behavior](../data/images/quick-breakdown-of-keep-default-na-behavior.png)

#### Image:
![Image of a CSF File](../data/images/read-csv-na-values.png)

```python
# 13.1 Default (keep_default_na=True)
erp_sales_report = pd.read_csv("../data/csv-files/ERP_sales_report_east_20260607_084932.csv",
                              sep=";", skiprows=3)
erp_sales_report
```

```python
# 13.2 keep_default_na=False
# Check the difference between keep_default_na=True and keep_default_na=False by looking at the outpot 13.1 & 13.2
erp_sales_report = pd.read_csv("../data/csv-files/ERP_sales_report_east_20260607_084932.csv",
                              sep=";", skiprows=3, keep_default_na=False)
erp_sales_report 
```

#### Note!
A common issue occurs when a column contains valid data (such as North America coded as "NA"), but pandas converts it to `NaN`. To stop this, set `keep_default_na=False` so pandas only flags what you explicitly tell it to flag.

```python
# 13.3 set keep_default_na=False and pass explicitly na_values
erp_sales_report = pd.read_csv("../data/csv-files/ERP_sales_report_east_20260607_084932.csv", sep=";",
                              skiprows=3, keep_default_na=False,
                               na_values=["unknown", "missing", "#QNAN", "1.#QNAN", "N/A", "#N/A", "NaN"])
erp_sales_report
```

### 14. `na_values` Parameter
The `na_values` parameter is used to specify custom strings, numbers, or markers that should be recognized as missing data (`NaN`). 

By default, pandas automatically interprets empty cells and common strings like `#N/A`, `N/A`, `NULL`, `NaN`, and null as missing values. If your dataset uses unique placeholders (like `?`, `-999`, or `Missing`), you must declare them using na_values.

#### Image:
![Quick Breakdown of keep_default_na Behavior](../data/images/quick-breakdown-of-keep-default-na-behavior.png)

![Image of a CSF File](../data/images/read-csv-na-values.png)

```python
# 14.1 Default
erp_sales_report = pd.read_csv("../data/csv-files/ERP_sales_report_east_20260607_084932.csv", sep=";",
                              skiprows=3)
erp_sales_report
```

```python
# 14.2 Pass a List
erp_sales_report = pd.read_csv("../data/csv-files/ERP_sales_report_east_20260607_084932.csv", sep=";",
                              skiprows=3, na_values=["unknown", "missing", "?", "-999"])
erp_sales_report
```

#### Note!
A common issue occurs when a column contains valid data (such as North America coded as "NA"), but pandas converts it to `NaN`. To stop this, set `keep_default_na=False` so pandas only flags what you explicitly tell it to flag.

```python
# 14.3 set keep_default_na=False and pass explicitly na_values
erp_sales_report = pd.read_csv("../data/csv-files/ERP_sales_report_east_20260607_084932.csv", sep=";",
                              skiprows=3, keep_default_na=False,
                               na_values=["unknown", "missing", "?", "-999", "#QNAN", "1.#QNAN", "N/A", "#N/A", "NaN"])
erp_sales_report
```

### 15. `na_filter` Parameter
`na_filter` parameter is a boolean flag (`True` by default) that controls whether pandas detects and filters missing value markers (like empty cells, NaN, NULL, N/A) into actual `NaN` values.

The core difference between `na_filter` and `keep_default_na` is their scope: `na_filter` is a master switch that turns missing value detection completely on or off, while `keep_default_na` controls which specific strings pandas considers to be missing values.

#### Image:m
![Image of a CSv File](../data/images/read-csv-na-values.png)

```python
# If na_filter=True, keep_default_na and na_values work!
erp_sales_report = pd.read_csv("../data/csv-files/ERP_sales_report_east_20260607_084932.csv", sep=";",
                              skiprows=3, na_values=["unknown", "missing", "?", "-999"], na_filter=True)
erp_sales_report
```

```python
# But if na_filter=False, keep_default_na and na_values don't work!
erp_sales_report = pd.read_csv("../data/csv-files/ERP_sales_report_east_20260607_084932.csv", sep=";",
                              skiprows=3, na_values=["unknown", "missing", "?", "-999"], na_filter=False)
erp_sales_report
```

### 18. `engine` Parameter
The `engine` parameter specifies the underlying parsing engine used to read and process your text files. Selecting the right engine allows you to balance execution speed, memory efficiency, and specific feature requirements.

#### 1. `engine='c'` (The Default)
* __Speed:__ Extremely fast because it runs on compiled `C` code.
* __Use Case:__ Best for standard, clean, or massive CSV datasets.
* __Limitations:__ It does not support regex delimiters, skipfooter, or advanced fallback behaviors.

#### 2. `engine='python'`
* __Speed:__ Significantly slower because it processes data in pure `Python`.
* __Use Case:__ Use this when you have complex or unconventional formatting.
* __Key Features:__ Supports multi-character regex delimiters (e.g., `sep='[;,]'`), skipping footers (`skipfooter=N`), and auto-detecting delimiters using Python’s built-in sniffer.

#### 3. `engine='pyarrow'`
* __Speed:__ The fastest engine option for large-scale data.
* __Use Case:__ Highly recommended for massive files where maximum I/O performance is critical.
* __Key Features:__ It uses Apache Arrow and is the only engine that natively supports multithreading to parallelize file reading.
* __Limitations:__ It requires the pyarrow library to be installed and may not support every minor formatting edge case.

### 19. `comment` Parameter
he comment parameter takes ___a single character___. Pandas ignores everything after this character on a line. If a line starts with it, the whole line is skipped. If it appears mid-line, pandas chops off the rest of that line. It parses only the data before the character.

#### Image:
![Image of a CSV File](../data/images/read-csv-skiprows1.png)

```python
# Commenting rows
heart_disease = pd.read_csv("../data/csv-files/heart-disease.csv", sep=";", comment="#")
heart_disease.head(3)
```
