## **Series**
---

### What a Series is
A Pandas Series is a one-dimensional labeled array capable of holding any data type, such as integers, floats, strings, or Python objects. Conceptually, it is like a single column in an Excel spreadsheet or a database table.

### Creating Series
To create a Pandas Series, use the `pd.Series()` constructor. Series can be build from various data sources like Python lists, NumPy arrays, dictionaries, or scalar values.

```python
# Creating an empty Series
my_series = pd.Series()
my_series
```

```python
# Creating a Series from a Python list
my_series = pd.Series(["Apple", "Banana", "Oringe"])
my_series
```

```python
# Use the index parameter to specify explicit labels for your data.
list_data = ["Apple", "Banana", "Oringe"]
my_series = pd.Series(list_data, index=["A", "B", "C"])
my_series
```

```python
# Creating a Series from a Python dictionary
dict_data = {"Apple": 5, "Banana": 3, "Oringe": 7}
my_series = pd.Series(dict_data)
my_series
```

```python
# Creating a Series frome a Numpy array
import numpy as np
np_data = np.array([1.4, 3.4, 8])
my_series = pd.Series(np_data)
my_series
```

```python
# Creating a Series from a scalar value
my_series = pd.Series(5, index=[x for x in "abcd"])
my_series
```

```python
# Creating a Series from a Python function
my_series = pd.Series(range(5))
my_series
```

## **DataFrame**
---

### What a DataFrame is
A Pandas DataFrame is a two-dimensional, tabular data structure in Python that organizes data into rows and columns, much like an Excel spreadsheet or a SQL database table. It can store different types of data such as numbers, text and dates across its columns.

### Creating DataFrame
To create a pandas DataFrame, use the `pd.DataFrame()` constructor. Pandas allows us to create a DataFrame from many data sources. We can create DataFrames directly from Python objects like lists and dictionaries or by reading data from external files like CSV, Excel or SQL databases.

```python
# Creating an empty DataFrame
my_dataframe = pd.DataFrame()
print(my_dataframe)
```

```python
# Creating a DataFrame from Python list
users_info = pd.DataFrame({"Name": ["Ali", "Mohammad", "Sadiq", "Hassan"], "Age": [22, 25, 18, 10]})
users_info
```

```python
# Use the index parameter to specify explicit labels for your data.
names = ["Ali", "Mohammad", "Sadiq", "Hassan"]
ages = [22, 25, 18, 10]
users_info = pd.DataFrame({"Name": names,"Age": ages}, index=["A", "B", "C", "D"])
users_info
```

```python
#  Creating a DataFrame from a list of lists (Row-Oriented)
data = [
    ['Ali', 22, 'Kabul'],
    ['Mohammad', 18, 'Bamyan'],
    ['Hussan', 25, 'Herat']
]
users_info = pd.DataFrame(data, columns=["Name", "Age", "City"])
users_info
```

```python
# Creating DataFrame from Python dictionary
data = {
    "Name":["Ali", "Mohammad", "Hassan"],
    "Age": [22, 18, 25],
    "City": ["Kabul", "Bamyan", "Herat"]
}
users_info = pd.DataFrame(data)
users_info
```

```python
# Creating a DataFrame from a list of dictionaries (Row-Oriented)
data = [
    {'Name': 'Ali', 'Age': 22, 'City': 'Kablu'},
    {'Name': 'Mohammad', 'Age': 18, 'City': 'Bamyan'},
    {'Name': 'Hassan', 'Age': 25, 'City': 'Herat'}
]

users_info = pd.DataFrame(data)
users_info
```

```python
# Creating a DataFrame from a Numpy array
data = {
    "A": np.array([1, 2, 3]),
    "B": np.array([4, 5, 6]),
    "C": np.array([7, 8, 9])
}
np_dataframe = pd.DataFrame(data)
np_dataframe
```
