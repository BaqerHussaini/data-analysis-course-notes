### Importing Data From SQL
`pandas.read_sql` is a convenience wrapper that loads the results of a SQL query or database table directly into a Pandas DataFrame. It automatically routes your request to either `read_sql_query` (if you pass a raw query string) or `read_sql_table` (if you pass a single table name)

Before you work with `pandas.read_sql`, install sqlalchemy and pymysql libraries. `pip install sqlalchemy pymysql`
