### 1. `sheet_name` Parameter
It determines which specific worksheet or worksheets to load from your Excel file. 

By default, it is set to `0`, which imports only the first sheet as a `DataFrame`.

#### Image:
![Image of a Excel File with Two Sheets](../data/images/read-excel-sheet-name.png)

```python
# 1.1 Load a Specific Sheet
# set sheet_name to a number like 0 or to sheet's name like sheet1
car_sales = pd.read_excel("../data/excel-files/car-sales.xlsx", sheet_name="car-sales-extended") 
car_sales.head(3)
```

```python
# 1.2 Load Multiple Specific Sheets
# set sheet_name to numbers like [0, 1] or sheet's names like ["sheet1", "sheet2"]
# Passing a list returns a dictionary where the keys are the sheet identifiers and the values are individual DataFrames:
car_sales = pd.read_excel("../data/excel-files/car-sales.xlsx", sheet_name=[0, 1])
car_sales_sheet1 = car_sales[0]
car_sales_extended = car_sales[1]
car_sales_sheet1.head(3)
```

### 6. `skipfooter` Parameter
It is used to ignore a specific number of rows at the very bottom (footer) of an Excel worksheet during import.

The primary difference with `read_csv` is that `skipfooter` works automatically in `read_excel`, but requires you to manually switch the parsing `engine` in `read_csv`.

#### Image:
![Image of an Excel File](../data/images/read-excel-skipfooter.png)

```python
laptop_sales = pd.read_excel("../data/excel-files/laptop-sales.xlsx", skiprows=[3], skipfooter=3)
laptop_sales
```

### 7. `usecols` Parameter
It allows you to select and load only specific columns into memory from a spreadsheet. Using this parameter improves speed and drastically reduces memory consumption when working with massive Excel workbooks.

#### Image:
![Image of an Excel File](../data/images/excel-file.png)

```python
# 7.1 Excel Column Letters (String)
# Both upper case letters and lower case letters work(a or A, b or B)
# 7.1.1 Specific column(s)
car_sales = pd.read_excel("../data/excel-files/car-sales.xlsx", usecols="A")
car_sales.head(3)
```

```python
car_sales = pd.read_excel("../data/excel-files/car-sales.xlsx", usecols="a, B, E") 
car_sales.head(3)
```

```python
# 7.1 Excel Column Letters (String)
# 7.1.2 Range of columns
car_sales = pd.read_excel("../data/excel-files/car-sales.xlsx", usecols="B:E")
car_sales.head(3)
```

```python
# 7.1 Excel Column Letters (String)
# 7.1.3 Mixed columns and ranges
car_sales = pd.read_excel("../data/excel-files/car-sales.xlsx", usecols="A:C, E")
car_sales.head(3)
```

```python
# 7.2 Column Names (List of Strings)
car_sales = pd.read_excel("../data/excel-files/car-sales.xlsx", usecols=["Make"])
car_sales.head(3)
```

```python
car_sales = pd.read_excel("../data/excel-files/car-sales.xlsx", usecols=["Make", "Price"])
car_sales.head(3)
```

```python
# 7.3 Column Integer Indices (List of Integers)
car_sales = pd.read_excel("../data/excel-files/car-sales.xlsx", usecols=[4])
car_sales.head(3)
```

```python
car_sales = pd.read_excel("../data/excel-files/car-sales.xlsx", usecols=[0, 4])
car_sales.head(3)
```

```python
# 7.4 Conditional Filtering (Callable/Function)
needed_cols = ["Make", "Color", "Odometer (KM)", "Price"]
car_sales = pd.read_excel("../data/excel-files/car-sales.xlsx", usecols=lambda x: x in needed_cols)
car_sales.head(3)
```

### 18. `engine` Parameter
It specifies the underlying Python library used to parse the Excel file. Because pandas does not natively read Excel binaries, it relies on third-party backend "engines" to convert spreadsheets into DataFrames.

By default (`engine=None`), pandas automatically selects the best engine based on your file extension. However, explicitly defining the engine is highly recommended to optimize speed, bypass file corruption errors, or handle specialized formats.


#### 1. `engine='openpyxl'`
* __Extension:__ .xlsx, .xlsm
* __Speed:__ Moderate.
* __Use Case:__ The standard default engine for modern, XML-based Excel files.
* __Limitation:__ High memory consumption on massive files.

#### 2. `engine='calamine'`
* __Extension:__ All formats (.xlsx, .xlsb, .xls, .ods)
* __Speed:__ Extremely fast (Rust-backed).
* __Use Case:__ The best choice for massive datasets needing maximum performance.
* __Limitation:__ Requires installing the python-calamine package.

#### 3. `engine='xlrd'`
* __Extension:__ .xlsSpeed: Fast.
* __Use Case:__ Strictly reserved for legacy, pre-2007 Excel files.
* __Limitation:__ No longer supports modern .xlsx files due to security updates.

#### 4. `engine='pyxlsb'`
* __Extension:__ .xlsb
* __Speed:__ Fast.
* __Use Case:__ Dedicated exclusively to Excel Binary Workbooks.
* __Limitation:__ Cannot read standard XML or legacy sheets.

#### 5. `engine='odf'`
* __Extension:__ .ods
* __Speed:__ Slow.
* __Use Case:__ Specifically for open-source OpenDocument sheets (LibreOffice/OpenOffice).
* __Limitation:__ Very narrow use case; poor performance on large files.

### 19. `comment` Parameter
It parameter is used to find a specific character in text cells and ignore any text that follows it on that line.

#### How It Works
* __Default Value:__ `comment=None` (it does not parse or drop any text based on a comment character).
* __Input:__ It accepts a single character string (like '#' or '//').
* __Behavior:__ If the specified character is found inside a cell, pandas ignores all data from that character to the end of that row line. If a line starts with the comment character, the whole line is skipped.

#### `comment` in `read_csv` vs `read_excel`
* __`comment` in `read_csv`:__ Completely strips the row. The row index skips it entirely, and the DataFrame is shorter.
* __`comment` in `read_excel`:__ It cuts off the text inside a cell. If it is in the middle of a cell, all text after it is erased. If it starts the row, the whole row becomes blank. However, it does not delete the row, so the row index and total count stay the same.

#### Image:
![Image of an Excel File](../data/images/read-excel-skipfooter.png)

```python
laptop_sales = pd.read_excel("../data/excel-files/laptop-sales.xlsx", comment="#")
laptop_sales
```
